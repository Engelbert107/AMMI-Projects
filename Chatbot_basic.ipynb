### I will be using the NLTK library in Python which is one of the best Python libraries for any task of natural language processing.

```python
from nltk.chat.util import Chat, reflections
```

### Let's create a list of queries and their responses for the chatbot.

```python
# pairs is a list of patterns and responses.

pairs = [
    [
        r"(.*)my name is (.*)",
        ["Hello %2, How are you today ?",]
    ],
    [
        r"(.*)help(.*) ",
        ["I can help you ",]
    ],
     [
        r"(.*) your name ?",
        ["My name is robot and I'm a chatbot .",]
    ],
    [
        r"how are you (.*) ?",
        ["I'm doing very well", "i am great !"]
    ],
    [
        r"sorry (.*)",
        ["Its alright","Its OK, never mind that",]
    ],
    [
        r"i'm (.*) (good|well|okay|ok)",
        ["Nice to hear that","Alright, great !",]
    ],
    [
        r"(hi|hey|hello|hola|holla)(.*)",
        ["Hello", "Hey there",]
    ],
    [
        r"what (.*) want ?",
        ["Make me an offer I can't refuse",]
        
    ],
    [
        r"(.*)created(.*)",
        ["Engelbert Tchinde created me using Python's NLTK library ","top secret ;)",]
    ],
    [
        r"(.*) (location|city) ?",
        ['Mbour Tripano, Senegal',]
    ],
    [
        r"(.*)raining in (.*)",
        ["No rain in the past 4 months here in %2","In %2 there is a 5% chance of rain",]
    ],
    [
        r"how (.*) health (.*)",
        ["Health is very important, but I am a computer, so I don't need to worry about my health ",]
    ],
    [
        r"(.*)(sports|game|sport)(.*)",
        ["I'm a very big fan of Hobby",]
    ],
    [
        r"who (.*) (Hobbies|Football)?",
        ["Eto'o"]
    ],
    [
        r"quit",
        ["Bye for now. See you soon :) ","It was nice talking to you. See you soon :)"]
    ],
    [
        r"(.*)",
        ['That is nice to hear']
    ],
]
```

```python
# default message at the start of chat
print("Hi, I'm the cleverprogrammer and I like to chat\nPlease type lowercase English language to start a conversation. \n\nType quit to leave ")
```

```python
# Create Chat Bot
chat = Chat(pairs, reflections)
```

```python
# Start conversation
chat.converse()
```
